```python
from pathlib import Path
from pyprojroot import here as get_project_root
import user_config
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
DATA_DIR = get_project_root() / Path('Data/AB_NYC_2019.CSV')
df = pd.read_csv(DATA_DIR)
```

```python
df_1 = df[["price","neighbourhood_group"]].copy()
df_1["neighbourhood_group"] =  df_1["neighbourhood_group"].astype("category")
df_1["price"] =  df_1["price"].astype("int64")
```

```python
neighbourhood_group = ['Manhattan', 'Brooklyn', 'Bronx',
             'Queens', 'Staten Island']

plt.figure(figsize=(20, 10))

for n in neighbourhood_group:
    # Subset to the neighbourhood_group
    subset = df_1[df_1['neighbourhood_group'] == n]
    
    # Draw the density plot
    sns.distplot(subset['price'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = n)
    
# Plot formatting
plt.legend(prop={'size': 16}, title = 'Price')
plt.title('Density Plot with Multiple Neighbourhood Group')
plt.xlabel('Price')
plt.ylabel('Density')
```
